```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
# Load the pre-trained Stable Diffusion model
def load_model():
    model_id = "runwayml/stable-diffusion-v1-5"  # Pre-trained model
    pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float32)  # Use float32 for CPU
    return pipe

# Generate an image from an input sentence
def generate_image(prompt, save_path="output_image.png"):
    # Load the Stable Diffusion pipeline
    pipe = load_model()

    # Generate the image
    with torch.no_grad():
        image = pipe(prompt).images[0]

    # Save the image
    image.save(save_path)
    print(f"Image saved to {save_path}")
```

```python
# Example usage
text_prompt = "A futuristic cityscape with flying cars during sunset"
generate_image(text_prompt)
```

```python
# Example usage
text_prompt = " a good college environment"
generate_image(text_prompt)
```
